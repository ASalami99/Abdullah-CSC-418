```python
#Importing necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dictionary storing user details
classmates = {
    "Abdullah": {"matno": 21120613126, "image": "img/Abdullah.jpg"},
    "Ada": {"matno": 21120612451, "image": "img/Ada.jpg"},
    "Chima": {"matno": 21120612567, "image": "img/Chima.jpg"},
    "Clement": {"matno": 21120612479, "image": "img/Clement.jpg"},
    "Daniel": {"matno": 21120612478, "image": "img/Daniel.jpg"},
    "Emeka": {"matno": 21120612449, "image": "img/Emeka.jpg"},
    "Ilyas": {"matno": 21120612448, "image": "img/Ilyas.jpg"},
    "Miracle": {"matno": 21120612475, "image": "img/Miracle.jpg"},
    "Tabitha": {"matno": 21120612496, "image": "img/Tabitha.jpg"},
    "Tireni": {"matno": 21120613127, "image": "img/Tireni.jpg"},
}

# Function to validate and display user details
def validate_user():
    while True:  # Infinite loop until correct login details
        # Collect user details
        username = input("Username: ").strip().capitalize()
        password = input("Matriculation Number: ").strip()

        if username in classmates:
            expected_password = str(classmates[username]["matno"])
            if password == expected_password:
                print("\nLogin successful! Welcome, {}.\n".format(username))
                
                # Different image enhancement techniques
                print("1. Adjust Brightness and Contrast")
                print("2. Sharpen Image")
                print("3. Remove Noise")
                print("4. Scale and Resize Image")
                print("5. Inverse Transformation")

                try:
                    response = int(input("Enter the number of the enhancement technique you would like to perform on your picture: "))
                    if response not in range(1, 6):
                        print("Sorry, please enter a number between 1 and 5.")
                        continue
                except ValueError:
                    print("Invalid input. Please enter a number.")
                    continue

                # Loading the image
                image = cv2.imread(classmates[username]["image"])      
                
                if response == 1:  # Adjusting Brightness and Contrast
                    plt.subplot(1, 2, 1)
                    plt.title("Original")
                    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) 

                    brightness = 5
                    contrast = 1.5
                    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)
                    
                    cv2.imwrite('img/contrast_image.jpg', image2)

                    plt.subplot(1, 2, 2)
                    plt.title("Brightness & Contrast")
                    plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
                    plt.show()

                elif response == 2:  # Sharpen Image
                    plt.subplot(1, 2, 1)
                    plt.title("Original")
                    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

                    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
                    sharpened_image = cv2.filter2D(image, -1, kernel)

                    cv2.imwrite('img/sharpened_image.jpg', sharpened_image)

                    plt.subplot(1, 2, 2)
                    plt.title("Sharpened")
                    plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
                    plt.show()

                elif response == 3:  # Remove Noise
                    plt.subplot(1, 2, 1)
                    plt.title("Original")
                    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

                    filtered_image = cv2.medianBlur(image, 15)

                    cv2.imwrite('img/Median-Blur.jpg', filtered_image)

                    plt.subplot(1, 2, 2)
                    plt.title("Median Blur")
                    plt.imshow(cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB))
                    plt.show()

                elif response == 4:  # Scale and Resize Image
                    plt.subplot(1, 2, 1)
                    plt.title("Original")
                    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

                    scaled_image = cv2.resize(image, None, fx=2, fy=2)

                    cv2.imwrite('img/Scaled.jpg', scaled_image)

                    plt.subplot(1, 2, 2)
                    plt.title("Scaled")
                    plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
                    plt.show()

                elif response == 5:  # Inverse Transformation
                    plt.subplot(1, 2, 1)
                    plt.title("Original")
                    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

                    inverse_image = 255 - image

                    cv2.imwrite('img/Inverse-Image.jpg', inverse_image)

                    plt.subplot(1, 2, 2)
                    plt.title("Inverse Color")
                    plt.imshow(cv2.cvtColor(inverse_image, cv2.COLOR_BGR2RGB))
                    plt.show()

                break  # leaving the loop after successful login
            else:
                print("WROONNNG Matriculation Number bruv! Please try again.")
        else:
            print("WROONNNG username. Please try again.")

# Calling the function
validate_user()
```
