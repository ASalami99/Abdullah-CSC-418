## Image Cropping

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('img/sst.jpg', 0)

#Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

cropped_img = img[450:700, 250:600]

#Plot the rotated image
plt.subplot(1, 2, 2)
plt.title("Cropped Image")
plt.imshow(cropped_img)

#Save rotated image
cv.imwrite('img/cropped_image.jpg', cropped_img)

cv.waitKey(0)
cv.destroyAllWindows()
```
